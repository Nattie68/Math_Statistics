```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator

alpha = 0.05

n = 3
m = 2

x_n = np.array([-1.11, -6.10, 2.42])
y_m = np.array([-2.29, -2.91])

sigma_x = 2
sigma_y = 1

delta = x_n.mean() - y_m.mean()

N = stats.norm(loc=0,
               scale=np.sqrt(sigma_x / n + sigma_y / m))

print(f"Значение delta: {delta}")
```

```python
q = N.ppf(1 - alpha)
print(f"Квантиль порядка 1 - alpha: {q}")
```

```python
p_value = N.sf(delta)
print(f"p-value: {p_value}")
```

Построить график мощности критерия

```python
_, axs = plt.subplots(1, 1, figsize=(10, 6))

axs.set_title("График мощности")
axs.set_xlabel("d")
axs.set_ylabel("W(d)")

d = np.linspace(0, 5, 100)

axs.plot(d, N.sf(-d + q),
         label="Мощность критерия",
         color="green")

axs.xaxis.set_minor_locator(AutoMinorLocator(10))
axs.yaxis.set_minor_locator(AutoMinorLocator(10))

plt.grid(which='minor', linestyle='--', linewidth=0.2)
plt.grid(which='major', linewidth=0.4)

plt.show()
```
