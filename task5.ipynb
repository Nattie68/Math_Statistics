### 5. Случайная величина распределена равномерно на отрезке $[\theta, 2\theta]$


#### f)
Сгенерируйте выборку объема n = 100 для некоторого значения параметра $\theta$. Вычислите указанные выше доверительные интервалы для доверительной вероятности 0.95





```python
import numpy as np
import random

theta = 1
n = 100
beta = 0.95

def p(x):
    if theta <= x <= 2 * theta:
        return 1 / theta
    else: 
        return 0

def F(x):
    if x <= theta:
        return 0
    elif x >= 2 * theta:
        return 1
    else:
        return x / theta - 1

def F_inv(y):
    return theta * (y + 1)

np.random.seed(42)
sample = [F_inv(random.random()) for _ in range(n)]
sample_max = max(sample)

sample
```

### Точный метод:

$$
P(\frac{max(x_i)}{\sqrt[n]{\frac{1 + \beta}{2}} + 1} < \theta < \frac{max(x_i)}{\sqrt[n]{\frac{1 - \beta}{2}} + 1}) = \beta
$$

```python
alpha = 1 - beta
lower_bound_exact = sample_max / (((1 + beta) / 2) ** (1 / n) + 1)
upper_bound_exact = sample_max / (((1 - beta) / 2) ** (1 / n) + 1)

print(f"Доверительный интервал: ({lower_bound_exact}, {upper_bound_exact})")
print(f"Длина: {upper_bound_exact - lower_bound_exact}\n")
```

### Асимптотический метод:

$$
P(\tilde{\theta}_1 - \frac{t_2 \cdot 2S\sqrt{n - 1}}{3n} < \theta < \tilde{\theta}_1 - \frac{t_1 \cdot 2S\sqrt{n - 1}}{3n}) = \beta
$$

```python
t1 = -1.96
t2 = 1.96
sample_mean = np.mean(sample)

theta_1 = 2 / 3 * sample_mean
s = np.sqrt(np.sum([(x - np.mean(sample)) ** 2 for x in sample]) / (n - 1))

lower_bound_asym = theta_1 - (t2 * 2 * s * np.sqrt(n - 1)) / (3 * n)
upper_bound_asym = theta_1 - (t1 * 2 * s * np.sqrt(n - 1)) / (3 * n)

print(f"Доверительный нтервал: ({lower_bound_asym}, {upper_bound_asym})")
print(f"Длина: {upper_bound_asym - lower_bound_asym}\n")
```

#### g)
Численно постройте бутстраповский доверительный интервал.

```python
n_bootstrap = 1000
theta_wave = (n + 1) / (2 * n + 1) * np.max(sample) # Оценка ОМП

bootstrap_diffs = []

for _ in range(n_bootstrap):
    bootstrap_sample = np.random.choice(sample, size=len(sample), replace=True)
    bootstrap_theta = (n + 1) / (2 * n + 1) * np.max(bootstrap_sample)
    bootstrap_diffs.append(bootstrap_theta - theta_wave)

bootstrap_diffs.sort()
t1 = bootstrap_diffs[int(n_bootstrap * (1 - beta) / 2)]
t2 = bootstrap_diffs[int(n_bootstrap * (1 + beta) / 2)]

lower_bound_boot = theta_wave - t2
upper_bound_boot = theta_wave - t1

print(f"Доверительный интервал: ({lower_bound_boot}, {upper_bound_boot})")
print(f"Длина: {upper_bound_boot - lower_bound_boot}\n")
```
