```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

alpha = 0.05
n1 = 139
n2 = 1000
df1 = n1 - 1
df2 = n2 - 1

F_left  = st.f.ppf(alpha/2, df1, df2)
F_right = st.f.ppf(1 - alpha/2, df1, df2)

lam = np.linspace(0.2, 5, 1000)

power = []
for l in lam:
    left_prob  = st.f.cdf(F_left / l, df1, df2) if l > 0 else 0
    right_prob = 1 - st.f.cdf(F_right / l, df1, df2)
    power.append(left_prob + right_prob)

plt.figure(figsize=(10, 6))
plt.plot(lam, power, linewidth=2.5, color='darkorange', label='$W(theta)$')

plt.xlim(0.2, 5)
plt.ylim(0, 1.2)
plt.xlabel('theta', fontsize=12)
plt.ylabel('$W(theta)$', fontsize=12)
plt.title('График мощности', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
